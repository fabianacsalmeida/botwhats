```python
media_apartamento = dados[dados['Tipo'] == 'Apartamento']['Valor'].mean() # Filter the DataFrame and calculate the mean of 'Valor' for 'Apartamento'
media_apartamento
```

```python
media_tipo = dados.groupby('Tipo')['Valor'].mean()
media_tipo
```

```python
media = dados.groupby('Tipo').mean(numeric_only=True)
media
```

```python
dftipo_preco =dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor', ascending=False)

dftipo_preco.plot(kind='bar', figsize=(10, 6))
```
